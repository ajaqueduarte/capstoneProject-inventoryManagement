```python
import os
import pandas as pd

# Directory paths
sales_mix_directory = 'sales_mix'
sales_mix_clean_directory = 'sales_mix_clean'

# Ensure the clean directory exists
os.makedirs(sales_mix_clean_directory, exist_ok=True)
file_paths = os.listdir(sales_mix_directory)
```

```python
def clean_and_save_file(file_name):
    # Load the data from the file
    file_path = os.path.join(sales_mix_directory, file_name)
    data = pd.read_excel(file_path, header=None)
    # the header row is the row that contains the column "name"
    header_row_idx = data.index[data[0] == 'Name'].tolist()

    # Check if we found the header row
    if header_row_idx:
        # Use the first occurrence of the header row to read the file again
        data = pd.read_excel(file_path, header=header_row_idx[0])
    else:
        # Handle the case where 'Name' is not found in the file
        raise ValueError(f"The column 'Name' was not found in the file {file_name}")
    #after finding the head, we can drop the other columns that we not gonna use. 
    #We are selecting only the columns that we are interested in
    cleaned_data = data[['Name', 'Gross Sales Total', 'Percent of Total Sales', 'Quantity Sold', 'Average Price']]
    cleaned_data.head()
    # On our data there are some rows that are not actual menu items, but rather categories or other non-menu items.
    # We can filter out these rows by removing rows with certain names.
    names_to_remove = [
        "Total", "Smoothie", "Classic", "Hi Protein", "Spirit", "Superfood Plus",
        "Superfood", "Refresh", "Combo", "Regular Combo", "Snack Combo",
        "Ingredients (Smoothie)", "NO BOOSTER", "No Froyo", "No Yogurt",
        "Secret/Feature", "Specialty", "Shot", "Condiments (Booster Ball)",
        "Booster Ball", "Booster Blends", "Grilled Fresh", "Lunch", "Breakfast",
        "Merchandise", "Retail", "Fresh Juice", "Condiments (Fresh Juice)",
        "Instructions", "Condiments (Inst.)", "Allergy:", "ALLERGY: tree nuts",
        "No Banana", "No Raspberry", "Special Prep", "whole wheat"
    ]

    # Filter out the rows with names in the list and any empty rows
    cleaned_data_filtered = cleaned_data[~cleaned_data['Name'].isin(names_to_remove)].dropna(how='all')

    # Construct the new file name and path
    date_part = ' '.join(file_name.split(' ')[2:4])
    new_file_name = date_part + '.csv'
    new_file_path = os.path.join(sales_mix_clean_directory, new_file_name) 
    # Save the cleaned data in new directory
    cleaned_data_filtered.to_csv(new_file_path, index=False)


```

```python
file_names = [f for f in os.listdir(sales_mix_directory) if f.endswith('.xlsx')]
# Process each file
for file_name in os.listdir(sales_mix_directory):
    if file_name.endswith('.xlsx'):
        try:
            clean_and_save_file(file_name)
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
```
